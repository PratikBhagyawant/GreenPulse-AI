```python
import pandas as pd

# List of object columns
cols_to_convert = [
    'pH',
    'Organic Carbon(OC)',
    'Available Sulphur(S)',
    'Available Zinc(Zn)',
    'Available Manganese(Mn)',
    'Available Copper(Cu)'
]

for col in cols_to_convert:
    df[col] = df[col].astype(str).str.strip()   # remove spaces
    df[col] = df[col].str.replace(r'[^0-9.]', '', regex=True)  # remove units/symbols
    df[col] = pd.to_numeric(df[col], errors='coerce')  # convert to float

# Check again
print(df.dtypes)
```

```python
import pandas as pd



# Clean column names
df.columns = df.columns.str.strip()

# Convert all object columns to numeric safely
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip()
    df[col] = df[col].str.replace(r'[^0-9.]', '', regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Fill missing values
df.fillna(df.mean(), inplace=True)
```

```python
import pandas as pd

def fertility_score(row):
    score = 0
    total = 12  # total parameters
    
    # pH (ideal range 6.0–7.5)
    if 6.0 <= row['pH'] <= 7.5:
        score += 1
        
    # Organic Carbon (>= 0.5 is minimum fertile)
    if row['Organic Carbon(OC)'] >= 0.5:
        score += 1
        
    # Nitrogen (>= 280 kg/ha)
    if row['Available\xa0Nitrogen(N)'] >= 280:
        score += 1
        
    # Phosphorus (>= 10 kg/ha)
    if row['Available Phosphorus(P)'] >= 10:
        score += 1
        
    # Potassium (>= 110 kg/ha)
    if row['Available Potassium(K)'] >= 110:
        score += 1
        
    # Sulphur (>= 10 ppm)
    if row['Available Sulphur(S)'] >= 10:
        score += 1
        
    # Zinc (>= 0.6 ppm)
    if row['Available Zinc(Zn)'] >= 0.6:
        score += 1
        
    # Boron (>= 0.5 ppm)
    if row['Available Boron(B)'] >= 0.5:
        score += 1
        
    # Iron (>= 4.5 ppm)
    if row['Available\xa0Iron(Fe)'] >= 4.5:
        score += 1
        
    # Manganese (>= 2.0 ppm)
    if row['Available Manganese(Mn)'] >= 2.0:
        score += 1
        
    # Copper (>= 0.2 ppm)
    if row['Available Copper(Cu)'] >= 0.2:
        score += 1
        
    # EC (Non-saline soil < 0.8 dS/m)
    if row['EC'] < 0.8:
        score += 1
        
    # 70% threshold
    if score >= 0.5 * total:
        return 1.0   # Fertile
    else:
        return 0.0   # Non-Fertile


# Apply to dataframe
df['Fertility'] = df.apply(fertility_score, axis=1)
```
